```python
#Removed the null columns Permanently

data.drop(['Intersection Street 1','Intersection Street 2','School or Citywide Complaint','Vehicle Type','Landmark','Taxi Company Borough','Taxi Pick Up Location','Bridge Highway Name','Bridge Highway Direction',
         'Road Ramp','Bridge Highway Segment','Garage Lot Name','Ferry Direction','Ferry Terminal Name'],axis=1,inplace=True)
```

Perform basic data exploratory analysis:

```python
# plot to show the number of null values in each column of the DataFrame after removing all blank columns

plt.figure(figsize=(10,6))  
data.isnull().sum().plot(kind="bar")
plt.title("Null values")
plt.xlabel("columns name")
plt.ylabel("Numbers")
plt.show()
```

```python
# Remove the records whose Closed Date values are null

data.dropna(subset=["Closed Date"], inplace= True) 
```

```python
#checking the null values in specific column

data["Closed Date"].isnull().sum()
```

Analyze the date column, and remove entries that have an incorrect timeline

```python
#convert closed/Created date arguments to datetime objects.

data['Closed Date'] = pd.to_datetime(data['Closed Date'])
data['Created Date'] = pd.to_datetime(data['Created Date'])
                                     
```

```python
#checking the type
data['Closed Date'].dtypes
data['Created Date'].dtypes
```

```python
# Calculate the time elapsed in closed and creation date

data['Time elapsed']= data['Closed Date']-data['Created Date']
data['Time elapsed']
```

```python
# Convert the calculated date to seconds to get better respresentation

data['Time elapsed'].dt.total_seconds()
```

```python
# Display the various types of complaints in each city ### Group by City and aggregate unique Complaint Types

complaints_by_city = data.groupby('City')['Complaint Type'].unique().reset_index()

complaints_by_city
```

```python
# Create a new DataFrame with cities as columns and complaint types in rows

df_new = data.pivot_table(index='Complaint Type', columns='City', aggfunc='size', fill_value=0)
df_new

```

```python
#Draw another chart that shows the types of complaints in each city in a single chart, where different colors show the different types of complaints

# Create a pivot table
complaint_counts = data.pivot_table(index='Complaint Type', columns='City', aggfunc='size', fill_value=0)

# Plotting
complaint_counts.plot(kind='bar', stacked=True, figsize=(12, 6), color=['lightblue', 'lightgreen', 'lightcoral'])

# Adding labels and title
plt.title('Types of Complaints in Each City')
plt.xlabel('Complaint Type')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=45)
plt.legend(title='City')
plt.tight_layout()  # Adjust layout to fit labels

# Show plot
plt.show()
```

```python
# Sort the complaint types based on the average Request_Closing_Time grouping them for different locations

avg_request_closing_time = data.groupby(['City', 'Complaint Type'])['Time elapsed'].mean()

sorted_complaint_types = avg_request_closing_time.groupby('City').apply(lambda x: x.sort_values(ascending = True))
sorted_complaint_types
```

```python
# Visualize the average of Request_Closing_Time

# Convert 'Time elapsed' from timedelta to hours
data['Time elapsed'] = data['Time elapsed'].dt.total_seconds() / 3600  # Convert to hours


average_request_closing_time = data.groupby('Complaint Type')['Time elapsed'].mean()

plt.figure(figsize=(9, 10))
average_request_closing_time.plot(kind='pie')

plt.title('Average Request Closing Time by Complaint Type')
plt.xlabel('Complaint Type')
plt.ylabel('Average Request Closing Time (in hours)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
